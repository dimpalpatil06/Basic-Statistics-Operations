```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency, chi2
```

```python
data = pd.DataFrame({
    'Satisfaction': ['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied'],
    'Smart Thermostat': [50, 80, 60, 30, 20],
    'Smart Light': [70, 100, 90, 50, 50],
    'Total': [120, 180, 150, 80, 70]
})
```

```python
#1. State the Hypotheses
print("Hypotheses")
print("H0: Device type and customer satisfaction are independent")
print("H1: Device type and customer satisfaction are associated\n")
```

```python
#2. Compute the Chi-Square Statistics
col_totals = [data['Smart Thermostat'].sum(), data['Smart Light'].sum()]
grand_total = sum(col_totals)

data['Expected Thermostat'] = data['Total'] * col_totals[0] / grand_total
data['Expected Light'] = data['Total'] * col_totals[1] / grand_total

# Chi-square contributions
chi2_stat = (((data['Smart Thermostat'] - data['Expected Thermostat'])**2 / data['Expected Thermostat']).sum() +
             ((data['Smart Light'] - data['Expected Light'])**2 / data['Expected Light']).sum())

print("Step 2: Chi-Square Statistic =", chi2_stat, "\n")
```

```python
#3. Determine the Critical Value
dof = (len(data)-1) * (2-1)  # (rows-1)*(cols-1)
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, dof)
print("Step 3: Degrees of Freedom =", dof)
print("Step 3: Critical Value at alpha=0.05 =", critical_value, "\n")
```

```python
#4. Make a decision
if chi2_stat > critical_value:
    print("Reject H0")
    print("Conclusion: There is a significant association between device type and customer satisfaction")
else:
    print("Fail to reject H0")
    print("Conclusion: There is no significant association between device type and customer satisfaction")
```
