```python
import numpy as np
from scipy import stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                       1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
# a. Build 99% Confidence Interval Using Sample Standard Deviation
```

```python
mean=np.mean(data)
std=np.std(data,ddof=1)
n=len(data)
print("Sample Mean:",mean)
print("Sample Standard Deviation:",std)
print("Size:",n)
```

```python
confidence_level = 0.99
alpha = 1 - confidence_level
df = n - 1  # degrees of freedom

t_critical = stats.t.ppf(1 - alpha/2, df)
print("t-critical value:", t_critical)
```

```python
margin_error = t_critical * (std / np.sqrt(n))
ci_lower = mean - margin_error
ci_upper = mean + margin_error

print(f"99% Confidence Interval (using sample std): ({ci_lower:.3f}, {ci_upper:.3f})")
```

```python
# b. Build 99% Confidence Interval Using Known Population Standard Deviation
population_std = 0.2
z_critical = stats.norm.ppf(1 - alpha/2)
print("z-critical value:", z_critical)

margin_error_pop = z_critical * (population_std / np.sqrt(n))
ci_lower_pop = mean - margin_error_pop
ci_upper_pop = mean + margin_error_pop

print(f"99% Confidence Interval (using population std): ({ci_lower_pop:.3f}, {ci_upper_pop:.3f})")
```
