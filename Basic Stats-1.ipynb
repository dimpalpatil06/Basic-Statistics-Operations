```python
# Plotting Histogram for each numerical column
# Analyzes skewness and outliers of each histogram
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram for Volume
sns.histplot(df_num['Volume'],bins=10,kde= True ,edgecolor="black")
df_num['Volume'].hist(bins=10,edgecolor="black")
plt.title("Histogram of Volume")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()

print("Skewness Of Volume: ", df_num["Volume"].skew())
Q1 = df_num['Volume'].quantile(0.25)
Q3 = df_num['Volume'].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers = df_num[(df_num['Volume'] < lower) | (df_num['Volume'] > upper)]['Volume']
print("Number of outliers in Volume:", len(outliers))
```

```python
# Box Plots for Numerical Variables
```

```python
for col in df_num.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df_num[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

    # IQR calculation
    Q1 = df_num[col].quantile(0.25)
    Q3 = df_num[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    # Outliers
    outliers = df_num[(df_num[col] < lower) | (df_num[col] > upper)][col]
    print(f"{col}  IQR = {IQR:.2f}, Outliers = {len(outliers)}")
```

```python
# Concept of Standardization (Z-Score Normalization)



```

```python
# Standardize the numerical columns using the formula: z=x-mu/sigma
df_standardized = df_num.copy()
for col in df_standardized.columns:
    mean = df_standardized[col].mean()
    std = df_standardized[col].std()
    df_standardized[col] = (df_standardized[col] - mean) / std

df_standardized.head()
```

```python
# Before and After Comparisons of the data distributions
for col in df_num.columns:
    plt.figure(figsize=(12,5))
    
    # Before standardization
    plt.subplot(1,2,1)
    sns.histplot(df_num[col], bins=20, kde=True, color='skyblue', edgecolor="black")
    plt.title(f"{col} - Before Standardization")
    
    # After standardization
    plt.subplot(1,2,2)
    sns.histplot(df_standardized[col], bins=20, kde=True, color='salmon', edgecolor="black")
    plt.title(f"{col} - After Standardization")
    
    plt.show()
```
