```python
# 1. State the Hypothesis
The restaurant owners claim that the weekly operating costs are higher than the model suggests.
Null hypothesis (Hâ‚€): The mean weekly cost is equal to the theoretical model.
W=1000+5X,X=600
W=1000+5âˆ—600=1000+3000=4000
Alternative hypothesis (Hâ‚): The mean weekly cost is higher than the theoretical model.
ð»1:ðœ‡>4000
```

```python
import math
from scipy.stats import norm

sample_mean = 3050
X = 600
theoretical_mean = 1000 + 5 * X
sigma = 5 * 25
n = 25
alpha = 0.05
```

```python
# 2. Calculate Test Statistic
z = (sample_mean - theoretical_mean) / (sigma / math.sqrt(n))
print(f"Test Statistic (z): {z:.2f}")
```

```python
# 3. Determine critical value (right-tailed test)
z_critical = norm.ppf(1 - alpha)
print(f"Critical Value (z_critical): {z_critical:.3f}")
```

```python
# 4. Make a decision
if z > z_critical:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"
print(f"Decision: {decision}")
```

```python
# 5. Conclusion
if z > z_critical:
    conclusion = "There is evidence that weekly operating costs are higher than the model suggests."
else:
    conclusion = "There is no evidence that weekly operating costs are higher than the model suggests."

print(f"Conclusion: {conclusion}")
```
